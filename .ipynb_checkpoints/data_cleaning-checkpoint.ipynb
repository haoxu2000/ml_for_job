# 1.导入库，数据

```python
import pandas as pd 

df = pd.read_csv('glassdoor_jobs.csv')
```

```python
df.head()
```

```python
df.columns
```

# 2. Salary Estimate
1.去除空值（值为-1）
2.删除多余符号，只留下数字
3.用年计算工资，或用小时计算工资，需要feature engineering
4.增加: avg_salary

```python
# 1.
df = df[df['Salary Estimate'] != '-1']
```

```python
# 2.
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
minus_K = salary.apply(lambda x: x.replace('$','').replace('K','').replace('k',''))
```

```python
# 3.
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary' in x.lower() else 0)

min_hr = minus_K.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary','').replace(':',''))
```

```python
# 4.
df['min_salary'] = min_hr.apply(lambda x: int(x.split('-')[0]))
df['max_salary'] = min_hr.apply(lambda x: int(x.split('-')[1]))
#hourly wage to annual
df['min_salary'] = df.apply(lambda x: x.min_salary*2 if x.hourly ==1 else x.min_salary, axis =1)
df['max_salary'] = df.apply(lambda x: x.max_salary*2 if x.hourly ==1 else x.max_salary, axis =1)
df['avg_salary'] = (df['min_salary']+df['max_salary'])/2
```

```python
df.head()
```

# 3. Job Description
分离出工作要求，增加几行：是否要求python，excel，r

```python
# python
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)

# r
df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
df.R_yn.value_counts()

#spark 
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df.spark.value_counts()

#aws 
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df.aws.value_counts()

#excel
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df.excel.value_counts()
```

# 4. Company name
只要文字，删除后面的评分

```python
df['company_txt'] = df.apply(lambda x: x['Company Name'] if x['Rating'] <0 else x['Company Name'][:-3], axis = 1)
df['company_txt'] = df.company_txt.apply(lambda x: x.replace('\n', ''))
```

# 5. Location
1.只提取州的位置
2.增加一行，看是否是总公司招聘
3.把洛杉矶转换为CA

```python
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
df.job_state.value_counts()

df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis = 1)
```

```python
# Fix state Los Angeles 
df['job_state']= df.job_state.apply(lambda x: x.strip() if x.strip().lower() != 'los angeles' else 'CA')
df.job_state.value_counts()
```

# 6. Job TItle
1.job title 换为小写
2.初级职位还是高级职位，还是不区分

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'
    
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
            return 'senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    else:
        return 'na'
```

```python
df['job_simp'] = df['Job Title'].apply(title_simplifier)
df.job_simp.value_counts()
```

```python
df['seniority'] = df['Job Title'].apply(seniority)
df.seniority.value_counts()
```

# 7. Job description length


```python
#  Job description length 
df['desc_len'] = df['Job Description'].apply(lambda x: len(x))
df['desc_len']
```

# 8.comptetitors
对竞争者计数

```python
df['num_comp'] = df['Competitors'].apply(lambda x: len(x.split(',')) if x != '-1' else 0)
```
